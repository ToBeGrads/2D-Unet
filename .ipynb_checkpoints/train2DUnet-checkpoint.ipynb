```python
%matplotlib inline

import os
import cv2
import json
import torch
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm import tqdm
from torch.utils.data import DataLoader

# Import your custom modules
from unet import UNet
from seg_dataset import get_patient_splits, MRISegDataset
from losses import FocalTverskyLoss
from ultralytics import YOLO

# ==========================================
# CONFIGURATION & PATHS
# ==========================================
ORIGINAL_IMAGE_DIR = Path("/home/jovyan/seg_data_modified_2d/images")
ORIGINAL_MASK_DIR  = Path("/home/jovyan/seg_data_modified_2d/masks")

YOLO_MODEL_PATH = "/home/jovyan/Localizer/runs/detect/mri_localizer11m/stn_polygon_model/weights/best.pt"

# New directories for the cropped dataset
CROPPED_IMAGE_DIR = Path("./Cropped_Dataset/images")
CROPPED_MASK_DIR  = Path("./Cropped_Dataset/masks")
CROPPED_IMAGE_DIR.mkdir(parents=True, exist_ok=True)
CROPPED_MASK_DIR.mkdir(parents=True, exist_ok=True)

MODEL_OUTPUT_DIR = Path("./Cropped_seg_final_training")
MODEL_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

CROP_SIZE = 96
BATCH_SIZE = 32
EPOCHS = 200
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(f"[INFO] Using execution device: {DEVICE}")

# ==========================================
# STEP 1: DATASET SPLITTING PER PATIENT
# ==========================================
# Note: We split based on the ORIGINAL masks so we know the assignments beforehand
train_p, val_p, test_p = get_patient_splits(str(ORIGINAL_MASK_DIR), val_fraction=0.1, test_fraction=0.1)
```

```python
# ==========================================
# STEP 2: YOLO CALIBRATION & CROPPING TO DISK
# ==========================================
print("\n[INFO] Initializing YOLO model...")
yolo_model = YOLO(YOLO_MODEL_PATH) if os.path.exists(YOLO_MODEL_PATH) else YOLO("yolo11m.pt")

all_image_paths = sorted(list(ORIGINAL_IMAGE_DIR.glob("*.png")))
valid_image_paths = []
vertical_gaps = []

print("[INFO] Scanning dataset to calculate ΔY and identify valid slices...")
for img_path in tqdm(all_image_paths, desc="Scanning"):
    results = yolo_model(img_path, verbose=False)[0]
    if results.boxes is not None and len(results.boxes) > 0:
        valid_image_paths.append(img_path)
        if len(results.boxes) >= 2:
            boxes = results.boxes.xyxy.cpu().numpy()
            boxes_sorted = boxes[np.argsort(boxes[:, 1])]
            y_center_top = (boxes_sorted[0, 1] + boxes_sorted[0, 3]) / 2
            y_center_bottom = (boxes_sorted[1, 1] + boxes_sorted[1, 3]) / 2
            vertical_gaps.append(y_center_bottom - y_center_top)

VERTICAL_DELTA_Y = np.mean(vertical_gaps) if len(vertical_gaps) > 0 else 35.0
print(f"✅ True anatomical vertical spacing (ΔY): {VERTICAL_DELTA_Y:.2f} px")

print(f"\n[INFO] Cropping and saving {len(valid_image_paths)} valid slices to disk...")
for img_path in tqdm(valid_image_paths, desc="Processing & Saving"):
    img = cv2.imread(str(img_path), cv2.IMREAD_GRAYSCALE)
    if img is None: continue
    img_h, img_w = img.shape
    
    mask_path = ORIGINAL_MASK_DIR / img_path.name
    mask = cv2.imread(str(mask_path), cv2.IMREAD_GRAYSCALE) if mask_path.exists() else None
    
    results = yolo_model(img_path, verbose=False)[0]
    target_boxes = results.boxes.xyxy.cpu().numpy()
    
    x1_real, y1_real = np.min(target_boxes[:, 0]), np.min(target_boxes[:, 1])
    x2_real, y2_real = np.max(target_boxes[:, 2]), np.max(target_boxes[:, 3])
    
    if len(target_boxes) >= 2:
        center_x, center_y = (x1_real + x2_real) / 2, (y1_real + y2_real) / 2
    else:
        if y1_real < img_h / 2:
            y1_combined = min(y1_real, y1_real + VERTICAL_DELTA_Y)
            y2_combined = max(y2_real, y2_real + VERTICAL_DELTA_Y)
        else:
            y1_combined = min(y1_real, y1_real - VERTICAL_DELTA_Y)
            y2_combined = max(y2_real, y2_real - VERTICAL_DELTA_Y)
        center_x, center_y = (x1_real + x2_real) / 2, (y1_combined + y2_combined) / 2
        
    crop_x1 = max(0, int(center_x - CROP_SIZE / 2))
    crop_y1 = max(0, int(center_y - CROP_SIZE / 2))
    crop_x1 = min(crop_x1, img_w - CROP_SIZE)
    crop_y1 = min(crop_y1, img_h - CROP_SIZE)
    crop_x2, crop_y2 = crop_x1 + CROP_SIZE, crop_y1 + CROP_SIZE
    
    # Save Image
    cv2.imwrite(str(CROPPED_IMAGE_DIR / img_path.name), img[crop_y1:crop_y2, crop_x1:crop_x2])
    
    # Save Mask (only if it exists in the original directory)
    if mask is not None:
        cv2.imwrite(str(CROPPED_MASK_DIR / img_path.name), mask[crop_y1:crop_y2, crop_x1:crop_x2])

# Dataset reduction stats
original_count = len(all_image_paths)
filtered_count = len(valid_image_paths)
print(f"\n[INFO] Done! Retained {filtered_count}/{original_count} slices ({(1 - filtered_count/original_count)*100:.2f}% reduction).")
```

```python
# ==========================================
# STEP 3: DATALOADERS & EVAL HELPERS
# ==========================================
# Point your ORIGINAL MRISegDataset class to the newly created CROPPED directories!
train_ds = MRISegDataset(str(CROPPED_IMAGE_DIR), str(CROPPED_MASK_DIR), train_p, augment=True, only_annotated=False)
val_ds   = MRISegDataset(str(CROPPED_IMAGE_DIR), str(CROPPED_MASK_DIR), val_p,   augment=False, only_annotated=False)
test_ds  = MRISegDataset(str(CROPPED_IMAGE_DIR), str(CROPPED_MASK_DIR), test_p,  augment=False, only_annotated=False)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)
val_loader   = DataLoader(val_ds,  batch_size=BATCH_SIZE, shuffle=False, num_workers=4)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)

class EarlyStopping:
    def __init__(self, patience=40, verbose=False, delta=0.001):
        self.patience, self.verbose, self.delta = patience, verbose, delta
        self.counter, self.best_score, self.early_stop = 0, None, False
        self.val_loss_min = float('inf')

    def __call__(self, val_loss, model, path):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
            self._save(val_loss, model, path)
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.verbose: print(f"  [EarlyStopping] Counter: {self.counter}/{self.patience}")
            if self.counter >= self.patience: self.early_stop = True
        else:
            self.best_score = score
            self._save(val_loss, model, path)
            self.counter = 0

    def _save(self, val_loss, model, path):
        if self.verbose: print(f"  [SAVED] Val loss optimized ({self.val_loss_min:.6f} -> {val_loss:.6f}). Saving.")
        torch.save(model.state_dict(), path)
        self.val_loss_min = val_loss

def evaluate(model, dataloader, criterion, device, desc="Val"):
    model.eval()
    total_loss, stn_inter, stn_union, rn_inter, rn_union = 0.0, 0, 0, 0, 0
    with torch.no_grad():
        for batch in tqdm(dataloader, desc=desc, leave=False):
            images, labels = batch['image'].to(device), batch['label'].to(device)
            outputs = model(images)
            total_loss += criterion(outputs, labels).item()

            preds = torch.argmax(outputs, dim=1)
            p_stn, t_stn = (preds == 1), (labels == 1)
            stn_inter += (p_stn & t_stn).sum().item()
            stn_union += (p_stn.sum() + t_stn.sum()).item()

            p_rn, t_rn = (preds == 2), (labels == 2)
            rn_inter += (p_rn & t_rn).sum().item()
            rn_union += (p_rn.sum() + t_rn.sum()).item()

    avg_loss = total_loss / len(dataloader)
    dice_stn = (2.0 * stn_inter) / (stn_union + 1e-6) if stn_union > 0 else float('nan')
    dice_rn  = (2.0 * rn_inter)  / (rn_union  + 1e-6) if rn_union  > 0 else float('nan')
    return avg_loss, dice_stn, dice_rn
```

```python
# ==========================================
# STEP 4: MODEL TRAINING LOOP
# ==========================================
best_model_path = os.path.join(MODEL_OUTPUT_DIR, "best_final_unet_model.pth")

features_config = [32, 64, 128, 256]
lr_config = 0.0005
weight_decay_config = 0.0001
dropout_config = 0.2
ce_w, ft_w = 0.1, 0.9

model = UNet(in_channels=1, num_classes=3, features=features_config, dropout_rate=dropout_config).to(DEVICE)

weights = torch.tensor([0.03, 0.5, 1.0], dtype=torch.float32).to(DEVICE)
criterion = FocalTverskyLoss(class_weights=weights, alpha=0.7, beta=0.3, gamma=0.75, ce_weight=ce_w, ft_weight=ft_w)
optimizer = torch.optim.AdamW(model.parameters(), lr=lr_config, weight_decay=weight_decay_config)
early_stop = EarlyStopping(patience=15, verbose=True)

print(f"\n🚀 Beginning final training run...")

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    
    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}", leave=False):
        images, labels = batch['image'].to(DEVICE), batch['label'].to(DEVICE)
        
        optimizer.zero_grad()
        loss = criterion(model(images), labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    train_loss /= len(train_loader)
    val_loss, dice_stn, dice_rn = evaluate(model, val_loader, criterion, DEVICE, desc=f"Epoch {epoch+1} Val")

    print(f"Epoch {epoch+1:03d} | Train: {train_loss:.4f} | Val: {val_loss:.4f} | STN: {dice_stn:.4f} | RN: {dice_rn:.4f}")

    early_stop(val_loss, model, best_model_path)
    if early_stop.early_stop:
        print("\n[EARLY STOPPING] Validation loss ceased decreasing. Halting.")
        break

print("\n[INFO] Evaluating best model on unseen Test Set...")
model.load_state_dict(torch.load(best_model_path))
test_loss, test_stn, test_rn = evaluate(model, test_loader, criterion, DEVICE, desc="Testing")

print("\n" + "="*50)
print("FINAL MODEL ASSESSMENT METRICS")
print("="*50)
print(f"  Test Loss: {test_loss:.4f} | STN Dice: {test_stn:.4f} | RN Dice: {test_rn:.4f}")
print("="*50)
```
